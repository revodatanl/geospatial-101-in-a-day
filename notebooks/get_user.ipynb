```python
import re

def get_username_from_email(email: str) -> str:
    """
    Extracts the username from an email address, removing any non-alphanumeric characters.

    Example:
    'melika.sajadian@domain.com' -> 'melikasajadian'
    """
    username = email.split('@')[0]
    cleaned = ''.join(re.findall(r'[a-zA-Z0-9]+', username))
    return cleaned
```
