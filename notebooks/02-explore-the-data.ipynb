```python
%sql
SELECT DISTINCT function
FROM geospatial.greenspaces.greenspace_site
ORDER BY function;
```

```python
%sql
SELECT COUNT(*)
FROM geospatial.greenspaces.greenspace_site
WHERE function IN ('Play Space', 'Playing Field', 'Public Park Or Garden');
```

```python
%sql
SELECT DISTINCT local_type
FROM geospatial.lookups.named_place
ORDER BY local_type;
```

```python
%sql
SELECT COUNT(*)
FROM geospatial.lookups.named_place
WHERE local_type IN ('Beach', 'Bus Station', 'Bus Station,Coach Station', 'Coach Station', 'Passenger Ferry Terminal', 'Railway Station') OR local_type LIKE '%Education%';
```

```python
%sql
SELECT COUNT(*)
FROM geospatial.lookups.named_place AS n
WHERE local_type IN ('Beach', 'Bus Station', 'Bus Station,Coach Station', 'Coach Station', 'Passenger Ferry Terminal', 'Railway Station') OR local_type LIKE '%Education%'
```

```python
%sql
SELECT COUNT(*)
FROM geospatial.lookups.named_place AS n
LEFT JOIN geospatial.lookups.boundary_line_ceremonial_counties As b
ON st_within(n.geometry, b.geometry)
WHERE local_type IN ('Beach', 'Bus Station', 'Bus Station,Coach Station', 'Coach Station', 'Passenger Ferry Terminal', 'Railway Station') OR local_type LIKE '%Education%'
```
